```python
import xarray as xr
import glob
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path ='/glade/scratch/wchapman/Reforecast/F006'

#find all files in directory                                                                                                                                                  
print('Training on')
train_file_names = sorted([f for f in glob.glob(path + "/train/*_500mb.nc", recursive=True)])
for f in train_file_names:
    print(f)

print('Validating on')
validate_file_names = sorted([f for f in glob.glob(path + "/validate/*_500mb.nc", recursive=True)])
for f in validate_file_names:
    print(f)
    
print('Testing on')
test_file_names = sorted([f for f in glob.glob(path + "/test/*_500mb.nc", recursive=True)])
for f in test_file_names:
    print(f)
    
All_file_names =train_file_names + validate_file_names +test_file_names 
```

```python



fcast = ['F'+f'{xx:03}' for xx in np.arange(0,126,6)]
# fcast = ['F060','F072']

checkvars =['IVT','T_p','Z_p','q_p','u_tr_p','v_tr_p']

for beans in fcast:
    path ='/glade/scratch/wchapman/Reforecast/'+beans

    #find all files in directory                                                                                                                                                  
    print('Training on')
    train_file_names = sorted([f for f in glob.glob(path + "/train/*_500mb.nc", recursive=True)])
    for f in train_file_names:
        print(f)

    print('Validating on')
    validate_file_names = sorted([f for f in glob.glob(path + "/validate/*_500mb.nc", recursive=True)])
    for f in validate_file_names:
        print(f)
    
    print('Testing on')
    test_file_names = sorted([f for f in glob.glob(path + "/test/*_500mb.nc", recursive=True)])
    for f in test_file_names:
        print(f)
    
    All_file_names =train_file_names + validate_file_names +test_file_names 


    

    for f in All_file_names:
        DAR =xr.open_dataset(f)
        print(np.sum(np.isnan(DAR)))
        for var in checkvars:
            print(f)
            IVT = DAR[var].data
#             print('SUM',np.sum(np.isnan(IVT)))
#             print('max:',np.max(IVT))
#             print('Im diving in !!!!! and I am fixing:',var)
            if np.sum(np.isnan(IVT)) > 0:
                print('in:', f)
                wherebad =np.where(np.isnan(IVT))
                print('Fixing here:',np.unique(wherebad[0]))
                IVT = DAR[var].data
                for ii in np.unique(wherebad[0]):
                    print('###################')
                    print('IM IN HERE!!!!! and I am fixing:',var)
                    print('###################')
                    print('HERE IS THE SUM BEFORE FIX: ',np.sum(np.isnan(IVT)))
                    inds = np.arange(0,IVT.shape[0])
                    inds_ii = np.where(wherebad[0]==ii)
                    avginds = np.delete(inds,np.unique(wherebad[0][inds_ii]))
                    IVT[ii,:,:] = np.mean(IVT[avginds.astype(int),:,:],axis=0) 
                    print('HERE IS THE SUM AFTER FIX: ',np.sum(np.isnan(IVT)))
                    print('replacing:',var)
                    DAR[var].data = IVT
            print('###################')
            print(np.sum(np.isnan(DAR)))
            print('###################')
        DAR.to_netcdf(f[:-3]+'_Clean.nc')
```
