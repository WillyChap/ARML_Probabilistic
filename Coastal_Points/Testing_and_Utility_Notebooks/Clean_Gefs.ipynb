```python
import xarray as xr
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
fcast = 'F036'

path ='/glade/scratch/wchapman/GEFS/'
train_file_names = sorted([f for f in glob.glob(path + fcast+"/WY*_"+fcast+".nc", recursive=True)])
for f in train_file_names:
    print(f)
```

```python
fcasts = ['F'+f'{x:03}' for x in range(0,126,6)]
for fcast in fcasts: 
    print(fcast)
    path ='/glade/scratch/wchapman/GEFS/'
    train_file_names = sorted([f for f in glob.glob(path + fcast+"/WY*_"+fcast+".nc", recursive=True)])

    for f in train_file_names:
        print(f)
        DAR =xr.open_dataset(f)
        IVT = DAR['IVT'].data
        inds = np.linspace(0,20,21)
        print(np.max(IVT))
        if np.max(IVT) > 5500:
            print('in:', f)
            wherebad =np.where(IVT==np.max(IVT))
            print('inds:', np.unique(wherebad[0]))
            IVT = DAR['IVT']
            for ii in np.unique(wherebad[0]):
                inds_ii = np.where(wherebad[0]==ii)
                avginds = np.delete(inds,np.unique(wherebad[1][inds_ii]))
                for bb in np.unique(wherebad[1][inds_ii]):
                    IVT[ii,bb,:,:] = np.mean(IVT[ii,avginds.astype(int),:,:],axis=0)
            DAR['IVT'] = IVT
        DAR.to_netcdf(f[:-3]+'_Clean.nc')
```
